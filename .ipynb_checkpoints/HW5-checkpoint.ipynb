```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.
from scipy.optimize import curve_fit

Data1 = np.genfromtxt('Data1.dat',delimiter = ',')
```

```python
t = Data1[0]
x = Data1[1]
y = Data1[2]

def ver(t,h,vy,g):
    return h + vy*t + (g*t**2)/2
def hor(t,vx):
    return vx*t

parm1, var1 = curve_fit(ver,t,y)
parm2, var2 = curve_fit(hor,t,x)
```

```python
h  = parm1[0]
vx = parm2[0]
vy = parm1[1]
g  = parm1[2]

print('h = ', h, '[m]')
print('v = ', np.sqrt(vx**2+vy**2),'[m/s]')
print('tan(theta) = ', vy/vx)
print('g = ', g, '[m/s/s]')

xx = hor(t,vx)
yy = ver(t,h,vy,g)

plt.plot(t,x,'r.')
plt.plot(t,xx)
plt.plot(t,y,'g.')
plt.plot(t,yy,'y')
plt.show()

plt.plot(x,y,'m.')
plt.plot(xx,yy,'c')
```

```python
#2.

pollution = np.genfromtxt('air_pollution.csv',delimiter = ',')
```

```python
month =  np.linspace(0,59,58)
ozone =  pollution[1:,1]
dust  =  pollution[1:,2]

from scipy.interpolate import UnivariateSpline

oz = UnivariateSpline(month, ozone)
du = UnivariateSpline(month, dust)
oz.set_smoothing_factor(0.0001)
du.set_smoothing_factor(1)

x = np.linspace(0,59,200)

plt.plot(month,ozone,'r.')
plt.plot(x,oz(x),'b')
plt.show()

plt.plot(month,dust,'g.')
plt.plot(x,du(x),'y')
plt.show()
```

```python
#3.

covid = np.genfromtxt('covid.csv',delimiter = ',')
```

```python
day = np.linspace(0,264,265)
confirmed = covid[1:,1]

from scipy import special

def f(x,a,b,c,d,e,f,g,h):
    return a*(special.erf(b*(x-c))+1)+d*(special.erf(e*(x-f))+1) + g*np.heaviside(x-h,1)*(x-h)**2

parm, var = curve_fit(f,day,confirmed,p0=[600,0,35,300,0,200,1,50])

x = np.linspace(0,264,1000)
plt.plot(day,confirmed,'r.')
plt.plot(x,f(x,*parm))
```

```python
#4.

Data = np.genfromtxt('Data.dat',delimiter = ',')
```

```python
t = Data[0]
x = Data[1]

def oscil(t,A,k,b,phi):
    m = 0.2 #[kg]
    ohm0 = np.sqrt(k/m)
    ohm = np.sqrt(ohm0**2-(b/(m*2))**2)
    return A*np.exp(-t*b/(m*2))*np.sin(ohm*t+phi)

parm, var = curve_fit(oscil,t,x,p0 = [0.5,1,0,np.pi/2])


k = parm[1]
b = parm[2]

print('k = ', k, '[N/m]')
print('b = ', b, '[N*s/m]')

plt.plot(t,x,'r.')
plt.plot(t,oscil(t,*parm))
```
