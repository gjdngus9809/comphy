```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#1.

#1-1

data1 = np.genfromtxt('datafolder/quiz5-1.txt',delimiter=' ')
t1 = data1[:,0]
y1 = data1[:,4]
poly1 = np.polyfit(t1,y1,1)
poly2 = np.polyfit(t1,y1,2)
poly4 = np.polyfit(t1,y1,4)
```

```python
t = np.linspace(0,1.4,50)

plt.plot(t1,y1,'o')
plt.plot(t,np.poly1d(poly1)(t),'r')
plt.plot(t,np.poly1d(poly2)(t),'g')
plt.plot(t,np.poly1d(poly4)(t),'y')
```

```python
#1-2
'''
since the expression for y is: y(t) = y0 + v0*t - (g*t**2)/2
quadratic fitting is the most accurate
also, if more data is collected,
the accuracy will increase in quadratic fitting while the accuracy goes down in the others.
'''
```

```python
#1-3
print('v0 = ',poly2[1])
print('g  = ',2*poly2[0])
```

```python
#2.
def f2(t,a,b):
    return a*t**2 + b*t

para1, var = curve_fit(f2,t1,y1)

t = np.linspace(0,1.4,50)
y2 = f2(t,*para1)

plt.plot(t1,y1,'o')
plt.plot(t,y2,'g')
print('v0 = ',para1[1])
print('g  = ',para1[0]*2)
```

```python
#3.

data3 = np.genfromtxt('datafolder/quiz5-3.txt',delimiter = ' ')
x3 = data3[:,0]
y3 = data3[:,1]

def f3(x,a1,a2,a3,b1,b2,b3,c1,c2,c3):
    return a1*np.exp(-(x-b1)**2/c1)+a2*np.exp(-(x-b2)**2/c2)+a3*np.exp(-(x-b3)**2/c3)

para2, var = curve_fit(f3,x3,y3,p0=[1,6,5,3,5,6.5,1,2,3])

x = np.linspace(0,10,200)

plt.plot(x3,y3,'o')
plt.plot(x,f3(x,*para2),'g')
```
