```python
import numpy as np
import matplotlib.pyplot as plt

#3
ey = np.array([0,1])
for a in np.arange(-1,1,0.2):
    A = np.array([[1,a],
                  [0,1]])
    v = np.dot(A,ey)
    plt.arrow(0,0,v[0],v[1], head_width = 0.1, color = 'r')
    
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```

```python
#4
B = np.array([[1,1],
              [1,-1]])
AB = np.array([[3,-1],
               [2,-2]])

Binv = np.linalg.inv(B)
A = np.dot(AB,Binv)
print(A)
```

```python
#5
#  = x
#    = y
# x + y = 40
# 4x+ 2y= 92

A = np.array([[1,1],
              [4,2]])
b = np.array([40,92])

sol = np.linalg.solve(A,b)
print('rabbit: ', int(sol[0]), ', chicken: ', int(sol[1]))
```
