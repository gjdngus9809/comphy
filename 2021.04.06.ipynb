```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import *
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,6,10,30])

pfit = np.polyfit(x,y,4) # y = 6.1 * x + - 7.7
```

```python
line = np.poly1d(pfit)
```

```python
xx = np.linspace(0,6,100)
yy = line(xx)

plt.plot(x,y,'.')
plt.plot(xx,yy)
```

```python
# curve_fit(f,xdata,ydata)
from scipy.optimize import curve_fit

def f(x,m,b):
    return m*x + b

parm, var = curve_fit(f,x,y) # can be used when f is non-polynomial.

yy1 = f(xx,parm[0],parm[1])

plt.plot(x,y,'.')
plt.plot(xx,yy1)
```

```python
x = np.array([-2,1,4,-1,3,-4])
y = np.array([-1,2,59,4,24,-53])

def f1(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x +d

parm1, var1 = curve_fit(f1,x,y)

xx  = np.linspace(-5,5,100)
yy2 = f1(xx,parm1[0],parm1[1],parm1[2],parm1[3])

plt.plot(x,y,'.')
plt.plot(xx,yy2)
```

```python
#data = np.genfromtxt('Data2.txt',delimiter = ',')
data = np.loadtxt('Data2.txt', dtype = None,delimiter=',')
```

```python
x = data[0]
y = data[1]

def f2(x,a,b,c):
    return a*np.sin(b*x+c)

parm2,cov = curve_fit(f2,x,y,p0=[0.5,2,np.pi/2]) # set initial condition
print(parm2)

yy = f2(x,parm2[0],parm2[1],parm2[2])

plt.plot(x,y,'.')
plt.plot(x,yy,'r')
```

```python
def v(t,v0,tau):
    return v0*(1 - np.exp(-t/tau))  # v = v0*(1-exp(-gamma*t))
```

```python
data1 = np.genfromtxt('oildata.dat',delimiter = ',')
```

```python
x = data1[0]
y = data1[1]

parm, cov = curve_fit(v,x,y,p0 = [5,0.1]) 

print(parm)

yy = v(x,parm[0],parm[1])

plt.plot(x,y,'.')
plt.plot(x,yy)
```

```python
#UnivariateSpline

from scipy.interpolate import UnivariateSpline

x = np.linspace(-3, 3, 50)
y = np.exp(-x**2) + 0.1 * np.random.randn(50)
plt.plot(x, y, 'ro', ms=5)

spl = UnivariateSpline(x, y)
xs = np.linspace(-3, 3, 1000)
plt.plot(xs, spl(xs), 'g', lw=3)

spl.set_smoothing_factor(0.5)
plt.plot(xs, spl(xs), 'b', lw=3)
plt.show()
```
