```python
ex = np.array([1,0])
ey = np.array([0,1])

plt.arrow(0,0,ex[0],ex[1],head_width=0.1, color = (0,0,1))
plt.arrow(0,0,ey[0],ey[1],head_width=0.1, color = (1,0,0))
plt.xlim(-2,2)
plt.ylim(-2,2)
```

```python
for alpha in np.arange(-1,1,0.2):
    v = ey + ex*alpha
    plt.arrow(0,0,v[0],v[1],head_width = 0.1)

plt.xlim(-2,2)
plt.ylim(-2,2)
```

```python
u = np.array([1,1])
for alpha in np.arange(-1,1,0.2):
    A = np.array([[1,alpha],
                  [0,alpha]])
    v = np.dot(A,u)
    
    plt.arrow(0,0,v[0],v[1],head_width =0.1)

plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()
```

```python
B = np.array([[1,1],
              [1,-1]])
AB = np.array([[3,-1],
               [2,-2]])

Binv = np.linalg.inv(B)
A = np.dot(AB,Binv)
print(A)
```
